### Problem 1: Unique Words in a Sentence

```python
# Take input
sentence = input("Enter sentence: ") 
# Convert sentence to list of words
words = sentence.split() 
# Convert list to set to remove duplicates
unique_words = set(words)
# Count unique words
count = len(unique_words) 
# Display result:
print("Unique Words count:", count)
print("Unique Words:", unique_words)
```

### Problem 2: Highest Salary from Employee Data

```python
# Dictionary storing employee name and salary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000 }
# Assume 1st employee has highest salary
highest_name = ""
highest_salary = 0
# Loop through dictionary
for name in employees:
    if employees[name] > highest_salary:
        highest_salary = employees[name]
        highest_name = name
# Display result:
print("Highest Salary:", highest_name, "-", highest_salary)
```

### Problem 3: Find Maximum and Minimum Values

```python
# List of numbers
numbers = [45, 22, 89, 10, 66]
# Assume 1st element as max and min
max_val = numbers[0]
min_val = numbers[0]
# Compare each number
for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num
# Display result:
print("List:", numbers)
print("Max:", max_val)
print("Min:", min_val)
```

### Problem 4: Count Products Above a Price Threshold

```python
# List of product prices
prices = [450, 1200, 899, 1500, 300]
count = 0
# Check each price
for price in prices:
    if price > 1000:
        count += 1
# Display result:
print("Products above 1000:", count)
```

### Problem 5: Calculate Attendance Percentage

```python
# Attendance list  "P" (Present) and "A" (Absent)
attendance = ["P", "P", "A", "P", "P"]
present_count = 0
# Count number of P
for day in attendance:
    if day == "P":
        present_count += 1
# Calculate percentage
total_days = len(attendance)
percentage = (present_count / total_days) * 100
# Display result:
print("Attendance Percentage:", percentage)
```

### Problem 7: Count Character Frequency

```python
# Input string
text = "python"
# Empty dictionary
freq = {}
# Count characters
for ch in text:
    if ch in freq:
        freq[ch] += 1
    else:
        freq[ch] = 1
# Display result:
print(freq)
```

### Problem 9: Check if a Key Exists in a Dictionary

```python
# Dictionary
employee = {
    "id": 101,
    "name": "Ravi",
    "salary": 75000}
# Key to check
key = "name"
# Check existence
if key in employee:
    print("Employee exists")
else:
    print("Employee does not exist")
```

### Problem 10: Calculate Average Marks

```python
# The list of marks
marks = [70, 75, 80, 65]
# Calculate average
total_marks = sum(marks)
count = len(marks)
average = total_marks / count

print("Average Marks:", average)
```
