### 1. User Login Check

```python
username= "admin"
password= "1234"
u= input("Enter username: ")
p= input("Enter password: ")
if u == username and p == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

### 2. Pass / Fail Analyzer

```python
marks = [45, 78, 90,  33, 60]
Count_pass = 0
Count_fail = 0
for mark in marks:
    if mark >= 50:
        Count_pass += 1
    else:
        Count_fail += 1
print("Total Students:", len(marks))
print("Pass Students:", Count_pass)
print("Fail Students:", Count_fail)     
```

### 3. Simple Data Cleaner

```python
names = [" Alice ", "bob", " CHARLIE "]
cleaned_list=[]
for name in names:
    name=name.strip()  # remove extra spaces
    name=name.lower()  # convert to lowercase
    cleaned_list.append(name)
print("Cleaned Names List:", cleaned_list)
```

### 4. Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"]
for msg in messages:
    length= len(msg)
    print(msg, " --> Length:", length)
    if length > 10:
        print("Long message found")
```

### 5. Error Message Detector

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = 0
for log in logs:
    if log == "ERROR":
        error_count += 1
print("Total ERROR Count:", error_count)
```
