### 1: Employee Performance Bonus Eligibility

```python
# employee names and scores
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
# find highest score
highest = 0
for score in employees.values():
    if score > highest:
        highest = score
# print employees having highest score
print("Top Performers Eligible for Bonus:")
for name, score in employees.items():
    if score == highest:
        print(name, "(Score:", score, ")")
```

###  2: Search Query Keyword Analysis

```python
# text given
text = "Buy mobile phone buy phone online"
# make text small letters
text = text.lower()
# remove punctuation
text = text.replace(",", "").replace(".", "")
words = text.split()
count = {}
# count each word
for word in words:
    if word in count:
        count[word] += 1
    else:
        count[word] = 1
# print only repeated words
for word in count:
    if count[word] > 1:
        print(word, ":", count[word])
```

###  3: Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid = []
# check each reading
for i in range(len(sensor_readings)):
    value = sensor_readings[i]
    # even number means valid
    if value % 2 == 0:
        valid.append((i, value))
print("Valid Sensor Readings:", valid)
```

### 4: Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
domain_count = {}
# count domains
for mail in emails:
    parts = mail.split("@")
    domain = parts[1]
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1
# total users
total = len(emails)
# print percentage
for domain in domain_count:
    percent = (domain_count[domain] / total) * 100
    print(domain, ":", percent, "%")
```

### 5: Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
# find average
total = sum(sales)
avg = total / len(sales)
limit = avg + (avg * 0.30)
# check spike days
for i in range(len(sales)):
    if sales[i] > limit:
        print("Day", i+1, ":", sales[i])
```

###  6: Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
count = {}
# count each id
for user in user_ids:
    if user in count:
        count[user] += 1
    else:
        count[user] = 1
# print duplicates only
for user in count:
    if count[user] > 1:
        print(user, "->", count[user], "times")
```
